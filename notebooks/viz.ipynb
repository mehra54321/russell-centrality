```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your final scores
df = pd.read_csv("../data/centrality_scores.csv")

# Bar chart: Degree Centrality
top_deg = df.nlargest(len(df), "Degree").set_index("Ticker")
plt.figure(figsize=(6,3))
top_deg["Degree"].plot.bar()
plt.title("Degree Centrality (ρ > 0.5)")
plt.ylabel("Degree")
plt.tight_layout()
plt.show()

# Bar chart: Closeness Centrality
top_clo = df.nlargest(len(df), "Closeness").set_index("Ticker")
plt.figure(figsize=(6,3))
top_clo["Closeness"].plot.bar()
plt.title("Closeness Centrality (ρ > 0.5)")
plt.ylabel("Closeness")
plt.tight_layout()
plt.show()
```
